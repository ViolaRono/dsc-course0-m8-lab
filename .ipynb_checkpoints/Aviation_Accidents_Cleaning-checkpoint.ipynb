```python
#load the aviation dataset
df = pd.read_csv("data/AviationData.csv", encoding="latin1", low_memory=False)
```

## Column by column description
1. *Event.Id*: Unique identifier for each accident event.  
2. *Investigation.Type*: Type of investigation (e.g., Accident or Incident).  
3. *Accident.Number*: Official accident reference number.  
4. *Event.Date*: Date when the accident occurred.  
5. *Location*: City or area where the accident happened.  
6. *Country*: Country where the accident occurred.  
7. *Latitude*: Geographic latitude of the accident location.  
8. *Longitude*: Geographic longitude of the accident location.  
9. *Airport.Code*: Code of the nearest airport.  
10. *Airport.Name*: Name of the nearest airport.  
11. *Injury.Severity*: Overall severity classification of the accident.  
12. *Aircraft.damage*: Level of damage to the aircraft.  
13. *Aircraft.Category*: Type or category of the aircraft.  
14. *Registration.Number*: Aircraft registration identifier.  
15. *Make*: Manufacturer of the aircraft.  
16. *Model*: Specific model of the aircraft.  
17. *Amateur.Built*: Indicates whether the aircraft was amateur-built.  
18. *Number.of.Engines*: Number of engines on the aircraft.  
19. *Engine.Type*: Type of engine used by the aircraft.  
20. *FAR.Description*: Regulatory classification of the aircraft.  
21. *Schedule*: Type of flight schedule (e.g., scheduled or non-scheduled).  
22. *Purpose.of.flight*: Purpose of the flight (e.g., personal, business).  
23. *Air.carrier*: Airline or operator of the aircraft.  
24. *Total.Fatal.Injuries*: Number of fatal injuries.  
25. *Total.Serious.Injuries*: Number of serious injuries.  
26. *Total.Minor.Injuries*: Number of minor injuries.  
27. *Total.Uninjured*: Number of people not injured.  
28. *Weather.Condition*: Weather conditions during the accident.  
29. *Broad.phase.of.flight*: Phase of flight when the accident occurred.  
30. *Report.Status*: Status or conclusion of the investigation.  
31. *Publication.Date*: Date when the report was published.

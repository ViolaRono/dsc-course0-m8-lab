## Exploratory Data Analysis

```python
df = pd.read_csv("clean_aviation_data.csv")

df.head()
```

Observation:

* The cleaned dataset was successfully loaded.

## Explore safety metrics across models/makes
- Remember that the client is interested in separate recommendations for smaller airplanes and larger airplanes. Choose a passenger threshold of 20 and separate the plane types. 

```python
#create small vs large aircraft
df['aircraft_size'] = df['total_passengers'].apply(
    lambda x: 'small' if x <= 20 else 'large'
)

df['aircraft_size'].value_counts()
```

Observation:
- The dataset is highly imbalanced, with 13,318 small aircraft and only 188 large aircraft.
- This indicates that the majority of recorded incidents involve smaller aircraft.
- As a result, analysis for large aircraft may be less reliable due to the significantly smaller sample size.

#### Analyzing Makes

Explore the human injury risk profile for small and larger Makes:
- choose the 15 makes for each group possessing the lowest mean fatal/seriously injured fraction
- plot the mean fatal/seriously injured fraction for each of these subgroups side-by-side

```python
# 1.Group data by aircraft size and make
# Then calculate the average fatal/serious injury rate for each group
make_risk = df.groupby(['aircraft_size', 'make'])['fatal_serious_rate'] \
              .mean() \
              .reset_index()

# 2: Filter for small aircraft and sort by lowest injury risk
# We select the top 15 safest manufacturers for small aircraft
small_safe = make_risk[make_risk['aircraft_size'] == 'small'] \
    .sort_values(by='fatal_serious_rate') \
    .head(15)

# 3: Filter for large aircraft and sort by lowest injury risk
# We select the top 15 safest manufacturers for large aircraft
large_safe = make_risk[make_risk['aircraft_size'] == 'large'] \
    .sort_values(by='fatal_serious_rate') \
    .head(15)

# Display results for both categories
small_safe, large_safe
```

Observation:*
- For small aircraft, even the safest manufacturers still have noticeable fatal/serious injury rates, ranging approximately from 0.16 to 0.32.
- This suggests that accidents involving small aircraft tend to result in higher injury risk overall.
- For large aircraft, several manufacturers show a fatal/serious injury rate of 0.00, indicating no recorded severe injuries in those cases.
- This may suggest that large aircraft are generally safer, although the small sample size (fewer large aircraft records) should be considered when interpreting these results.

**Distribution of injury rates: small makes**

Use a violinplot to look at the distribution of the fraction of passengers serious/fatally injured for small airplane makes. Just display makes with the ten lowest mean serious/fatal injury rates.

```python
#Step 1: get top 10 safest small aircraft makes
top_small_10 = small_safe.head(10)['make']

# Step 2: filter dataset for only those makes
small_df = df[(df['aircraft_size'] == 'small') & (df['make'].isin(top_small_10))]

# Step 3: plot violin plot
plt.figure(figsize=(10,6))

sns.violinplot(
    data=small_df,
    x='make',
    y='fatal_serious_rate',
    color = 'violet'
)

plt.xticks(rotation=45)
plt.title("Distribution of Fatal/Serious Injury Rates (Small Aircraft)")
plt.show()
```

Observation:

* Injury rates for small aircraft vary across manufacturers, with most values concentrated at lower levels (0.0–0.3).
* However, the wide spread and long tails show that severe injuries still occur in some cases.
* This suggests that small aircraft have more variability and higher risk in accident outcomes.

**Distribution of injury rates: large makes**

Use a stripplot to look at the distribution of the fraction of passengers serious/fatally injured for large airplane makes. Just display makes with the ten lowest mean serious/fatal injury rates.

```python
#1. get top 10 safest large aircraft makes (lowest injury rates)
large_top10 = large_safe.head(10)['make']

#2. filter dataset for these makes
large_data = df[df['make'].isin(large_top10) & (df['aircraft_size'] == 'large')]

#3. plot stripplot
plt.figure(figsize=(10,6))
sns.stripplot(
    data=large_data,
    x='make',
    y='fatal_serious_rate',
    color='red',
    jitter=True
)

plt.xticks(rotation=45)
plt.title('Distribution of Injury Rates for Large Aircraft Makes')
plt.show()
```

Observation:
- ⁠Most large aircraft manufacturers show injury rates clustered very close to 0.0, indicating that severe injuries are rare in many incidents.
- However, a few manufacturers (notably Boeing and Embraer) display higher variability, with some cases reaching very high injury rates.
- This suggests that while large aircraft are generally safer, when severe incidents occur, they can still result in significant injuries.
- ⁠Compared to small aircraft, the overall distribution is still more concentrated at low values, indicating better consistency in safety outcomes.
- ⁠The presence of outliers highlights the importance of considering both average risk and extreme cases in the analysis.

### Exploring Other Variables
- Investigate how other variables effect aircraft damage and injury. You must choose **two** factors out of the following but are free to analyze more:

- Weather Condition
- Engine Type
- Number of Engines
- Phase of Flight
- Purpose of Flight

For each factor provide a discussion explaining your analysis with appropriate visualization / data summaries and interpreting your findings.

```python
#1.Injury rates by weather conditon
weather_analysis = df.groupby('weather_condition')['fatal_serious_rate'].mean().sort_values()
weather_analysis
```

```python
weather_analysis.plot(kind='barh', figsize=(8,5), color = "#09637E", edgecolor = "black")
plt.title('Injury Rates by Weather Condition')
plt.xlabel('Fatal/Serious Injury Rate')
plt.show()
```

Observation:
* Injury rates are highest under IMC (Instrument Meteorological Conditions).
* IMC conditions involve poor visibility, which increases the likelihood of severe accidents.
* VMC (Visual Meteorological Conditions) shows significantly lower injury rates, indicating safer outcomes in clear weather.
* The “unknown” category also shows relatively high injury rates, though this may be due to missing or inconsistent data.
* Overall, adverse weather conditions are associated with more severe accident outcomes.
* This suggests that weather plays a critical role in aviation safety and injury severity.

```python
#2. Injury rates by no of engines
engine_analysis = df.groupby('number_of_engines')['fatal_serious_rate'].mean().sort_values()
engine_analysis
```

```python

plt.figure(figsize=(8,5))
colors = ['#6BAED6', '#3182BD', '#08519C']  # soft blues
box = df.boxplot(
    column='fatal_serious_rate',
    by='number_of_engines',
    patch_artist=True
)

for patch, color in zip(box.artists, colors):
    patch.set_facecolor(color)

plt.title('Injury Rates by Number of Engines')
plt.suptitle('')
plt.xlabel('Number of Engines')
plt.ylabel('Fatal/Serious Injury Rate')

plt.show()
```

Observatio:
* Aircraft with fewer engines (especially single-engine) tend to have higher and more variable injury rates.
* Multi-engine aircraft generally show lower fatal and serious injury rates.
* The presence of multiple engines may provide added safety through redundancy in case of failure.
* Overall, aircraft with more engines appear to have safer and more stable injury outcomes.
